```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
```

```python
exam.assign(total = exam['front']+ exam['back'] + exam['design'])
```

```python
exam.assign(
    total = exam['front'] + exam['back'] +exam['design'],
    mean = (exam['front'] +exam['back']+exam['design']/3)
)
```

```python
exam.assign(test = np.where(exam['design'] >= 60, 'pass', 'fail'))
```

```python
exam.assign(total = exam['front'] + exam['back'] +exam['design'])\
            .sort_values('total')
```

```python
exam.assign(total = exam['front'] + exam['back'] +exam['design'],
            mean = lambda x:x['total']/3)
```
